```python
from scipy import signal 
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
fs=1000
step=1/fs
t=np.arange(-10,10,step)
x=np.cos(4*np.pi*t)*np.exp(-0.5*t)*(t>=-3)-np.cos(4*np.pi*t)*np.exp(-0.5*t)*(t>=5)
plt.plot(t,x)
```

```python
k=np.arange(len(x))
X=step*sp.fft(x)*np.exp(1j*(-10)*fs*2*np.pi*k/len(x))
```

```python
A=np.absolute(np.fft.fftshift(X)) #amplitude
```

```python
P=np.angle(np.fft.fftshift(X)) #phase
```

```python
F=np.linspace(-fs/2,fs/2,len(X)) #frequency
```

```python
length=len(X)//2+50
```

```python
plt.plot(F[-length:length],A[-length:length])
plt.title ('Amplitude spectrum')
```

```python
plt.plot(F[-length:length],P[-length:length])
plt.title ('Phase spectrum')
```

```python
n=np.arange(-10,40)
x=0.9**n*np.cos(0.1*np.pi*n)*(n>=-1)
plt.stem(n,x)
```

```python
k=np.arange(len(x))
X=sp.fft(x)*np.exp(1j*-10*2*np.pi*k/len(x))
```

```python
A=np.absolute(X) #amplitude
```

```python
P=np.angle(X) #phase
```

```python
O=np.linspace(0,2*np.pi,len(X)) #omega
```

```python
plt.stem(O,A)
plt.title ('Amplitude spectrum')
```

```python
plt.stem(O,P)
plt.title('Phase spectrum')
```
